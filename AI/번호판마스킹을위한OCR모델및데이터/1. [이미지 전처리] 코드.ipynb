## 1. 이미지 resize

### - 이미지를 640 x 640으로 resize

```python
import cv2
import os
import matplotlib.pyplot as plt
```

```python
images_path = "files/Custom Dataset/images"

# 해당 폴더의 모든 이미지를 리스트로 저장
images = os.listdir(images_path)
```

```python
# 모든 이미지에 대하여 실행함
for image in images:
    
    img_path = images_path + '/' + image
    
    # 해당 경로에 해당하는 이미지를 읽어옴
    img_temp = cv2.imread(img_path)
    
    # 이미지를 640 x 640의 사이즈를 갖도록 조정 
    resized = cv2.resize(img_temp, (640,640))
    
    # 사이즈를 조정한 이미지를 다시 저장
    cv2.imwrite(f'files/Custom Dataset/resized/{image}', resized)
```

```python
# 샘플로 위에서 저장된 이미지 중 하나를 확인함
sample = cv2.imread("files/Custom Dataset/resized/data (13).jpg")
plt.imshow(sample)
plt.show()
```
